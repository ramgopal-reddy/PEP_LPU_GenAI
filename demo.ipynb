```python
# Q1=df_values['Area (mean)']['25%']
# Q3=df_values['Area (mean)']['75%']
Q1=df['Area (mean)'].quantile(.25)
Q3=df['Area (mean)'].quantile(.75)
```

```python
df['Area (mean)'].quantile(.25)
```

```python
Q1
```

```python
IQR=Q3-Q1
```

```python
LowerBound=Q1-1.5*(IQR)
UpperBound=Q3+1.5*(IQR)
```

```python
print(LowerBound)
print(UpperBound)
```

```python
print("LowerBound: " , LowerBound , "\nUpperBound: " , UpperBound)
```

```python
outlierValues=[]

for i in df['Area (mean)']:
    if i > UpperBound or i < LowerBound:
        outlierValues.append(i)
```
